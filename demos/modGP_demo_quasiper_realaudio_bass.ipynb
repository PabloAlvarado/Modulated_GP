```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile as wav
%matplotlib inline
import modgp
import GPflow
```

```python
def logistic(x):
    return 1./(1+ np.exp(-x))
```

```python
[sf, yaudio] = wav.read('test1_iso_sound_mono_8kHz.wav')
yaudio = (yaudio / (np.max(np.abs(yaudio))))[:,None]
#yaudio = yaudio[:,None]
sper = 1. / sf
N = np.size( yaudio)

t = np.linspace(0,(N-1)*(sper),N).reshape(-1, 1)
plt.plot(t, yaudio)


```

```python

import matplotlib.pyplot as plt
import plotly.plotly as py
import numpy as np
# Learn about API authentication here: https://plot.ly/python/getting-started
# Find your api_key here: https://plot.ly/settings/api



Fs = 8000.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t1 = np.arange(0,1,Ts) # time vector

ff = 100;   # frequency of the signal
y = np.sin(2*np.pi*ff*t)

y = yaudio[:,0]
t1 = t

n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(n/2)] # one side frequency range



Y = np.fft.fft(y)/n # fft computing and normalization
Y = Y[range(n/2)]

fig, ax = plt.subplots(2, 1)
ax[0].plot(t1,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].stem(frq,abs(Y)/ (np.max(np.abs(Y))),'r') # plotting the spectrum
ax[1].set_xlim([100, 300])
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[1].set_xticks([98.00, 103.83,110.00,116.54,123.47, 130.81,138.59,146.83,2*130.81])


#103.83 	332.29
#A2	110.00 	313.64
# A#2/Bb2  	116.54 	296.03
#B2	123.47 	279.42
#C3	130.81 	263.74
# C#3/Db3  	138.59 	248.93
#D3	146.83 	234.96
# D#3/Eb3  	155.56
```

```python
X = t

k1 = GPflow.kernels.Matern52(input_dim=1, lengthscales=10.)
k2 = GPflow.kernels.Matern52(input_dim=1, lengthscales=0.05, variance=10.)
frec = 130.81
k_i = GPflow.kernels.Cosine(input_dim = 1, variance=1.0,  lengthscales=1./(2*np.pi*frec))
k_j = GPflow.kernels.Cosine(input_dim = 1, variance=0.5,  lengthscales=1./(2*2*np.pi*frec))
k_l = GPflow.kernels.Cosine(input_dim = 1, variance=0.25, lengthscales=1./(3*2*np.pi*frec))
k3 = k1*(k_i + k_j + k_l)
```

```python
K2 = k2.compute_K_symm(X)
K3 = k3.compute_K_symm(X)
```

```python
noise_var = 0.001
np.random.seed(1)
f = np.random.multivariate_normal(np.zeros(X.shape[0]), K3).reshape(-1, 1)
g = np.random.multivariate_normal(np.zeros(X.shape[0]), K2).reshape(-1, 1)
mean = f * logistic(g)
y = mean + np.random.randn(*mean.shape) * np.sqrt(noise_var)
```

```python
plt.plot(X, f, 'b')
plt.twinx()
plt.plot(X, logistic(g), 'g', lw=2)
```

```python
plt.figure()
plt.plot(t, yaudio)
plt.plot(X, y)
```

```python
yaudio =yaudio.astype(np.float64)
m = modgp.ModGP(X, yaudio, k3, k2, X[::2].copy())
```

```python
m.kern1.fixed = True
m.kern2.fixed = True
m.likelihood.noise_var = noise_var
m.likelihood.noise_var.fixed = True
```

```python
m.optimize(disp=1, maxiter = 2000)
```

```python
muf, var = m.predict_f(X)
plt.figure()
plt.plot(X, muf, 'b')
plt.plot(X, muf + 2*np.sqrt(var), 'b--')
plt.plot(X, muf - 2*np.sqrt(var), 'b--')
plt.twinx()
mug, var = m.predict_g(X)
plt.plot(X, logistic(mug), 'g', lw=2)
plt.plot(X, logistic(mug + 2*np.sqrt(var)), 'g--')
plt.plot(X, logistic(mug - 2*np.sqrt(var)), 'g--')
```
